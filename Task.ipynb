Predictive Modeling

```python
x=df[['sd', 'Q25', 'IQR', 'sp.ent', 'meanfun']]
y=df['label_num']
```

```python
y
```

Splitting into train and test

```python
from sklearn.model_selection import train_test_split 
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=42)
```

Logistic Regression

```python
from sklearn.linear_model import LogisticRegression 
lorg=LogisticRegression(random_state=2) 
lorg.fit(x_train,y_train)
```

```python
y_pred=lorg.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix 
cm=confusion_matrix(y_test,y_pred)
cm
```

Accuracy

```python
from sklearn.metrics import accuracy_score 
print(accuracy_score(y_test,y_pred))
```

```python
from sklearn import metrics 
print(metrics.confusion_matrix(y_test, y_pred)) 
confusion = metrics.confusion_matrix(y_test, y_pred) 
TP = confusion[1, 1]
TN = confusion[0, 0] 
FP = confusion[0, 1] 
FN = confusion[1, 0]
ACC = (TP + TN) / float(TP + TN + FP + FN)
RE = TP / float(TP + FN)
SPE = TN / float(TN + FP)
PRE = TP / float(TP + FP)
print('Accuracy is=>',(TP + TN) / float(TP + TN + FP + FN)) 
print('Recall is=>',TP / float(TP + FN))
print('specificity is=>',TN / float(TN + FP)) 
print('precision is=>',TP / float(TP + FP))
```

Metrics of Logistic Regression

```python
table=pd.DataFrame({'Metric':['Accuracy', 'Recall', 'Specificity', 'Precision'], 'Value':[ACC, RE, SPE, PRE]})
table
```

ROC and AUC curve

```python
from sklearn.metrics import roc_auc_score 
from sklearn.metrics import roc_curve, auc
from sklearn.linear_model import LogisticRegression 
log_reg=LogisticRegression() 
log_reg.fit(x_train,y_train)
y_pred=log_reg.predict_proba(x_test)[:,1]
fpr, tpr, thresholds = metrics.roc_curve(y_test, y_pred) 
roc_auc = auc(fpr, tpr)
plt.plot(fpr, tpr, 'b', label='AUC = %0.2f'%roc_auc)
plt.legend(loc='lower right')
plt.plot([0, 1], [0, 1], 'r--')
plt.title('ROC curve')
plt.xlabel('False Positive Rate (1 - Specificity)') 
plt.ylabel('True Positive Rate (Sensitivity)')
```

Comparing with Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
gnb=GaussianNB()
y_pred = gnb.fit(x_train,y_train).predict(x_test)
```

```python
from sklearn import metrics 
print(metrics.confusion_matrix(y_test, y_pred)) 
confusion = metrics.confusion_matrix(y_test, y_pred) 
TP = confusion[1, 1]
TN = confusion[0, 0] 
FP = confusion[0, 1] 
FN = confusion[1, 0]
ACC = (TP + TN) / float(TP + TN + FP + FN)
RE = TP / float(TP + FN)
SPE = TN / float(TN + FP)
PRE = TP / float(TP + FP)
print('Accuracy is=>',(TP + TN) / float(TP + TN + FP + FN)) 
print('Recall is=>',TP / float(TP + FN))
print('specificity is=>',TN / float(TN + FP)) 
print('precision is=>',TP / float(TP + FP))
```
